- Handle the outliers in the dataset.

```python
import numpy as np

def detect_outliers_z_score(data, threshold=3):
    outliers = []
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers

year_outliers = detect_outliers_z_score(df["year"])
passengers_outliers =  detect_outliers_z_score(df["passengers"])

print("Outliers in year:", year_outliers)
print("Outliers in passengers:", passengers_outliers)
```

#### => Their is no outlier in this dataset
